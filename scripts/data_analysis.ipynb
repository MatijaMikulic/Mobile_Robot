```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Učitavanje podataka iz Excel datoteke
file_path = 'newest_results.xlsx'  # zamijeni putanjom tvoje datoteke
data = pd.read_excel(file_path)

# Osiguranje da su relevantni stupci numerički
numeric_columns = ['pogreška (x)', 'pogreška (y)', 'rel. Pogreška (x)', 'rel. Pogreška (y)',
                   'nakon rot (x)', 'nakon rot (y)', 'početni kut', 'konačni kut',
                   'aps poč. pogreška', 'aps kon. pogreška']
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Osiguravanje da je stupac 'putanja' string
data['putanja'] = data['putanja'].astype(str)

# Uzmi apsolutne vrijednosti pogrešaka
data['pogreška (x)'] = data['pogreška (x)'].abs()
data['pogreška (y)'] = data['pogreška (y)'].abs()

# Uklanjanje NaN vrijednosti samo iz relevantnih stupaca za box plotove
data_clean = data.dropna(subset=['pogreška (x)', 'pogreška (y)'])

# 1. Box plot za apsolutne pogreške u X osi po segmentima
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_clean['putanja'], y=data_clean['pogreška (x)'], palette="Set3")
plt.title('Kutijasti dijagram za apsolutne pogreške u X osi po segmentima')
plt.xlabel('Segment Putanje')
plt.ylabel('Apsolutna pogreška u X osi (cm)')
plt.grid(True)
plt.show()

# 2. Box plot za apsolutne pogreške u Y osi po segmentima
plt.figure(figsize=(10, 6))
sns.boxplot(x=data_clean['putanja'], y=data_clean['pogreška (y)'], palette="Set2")
plt.title('Kutijasti dijagram za apsolutne pogreške u Y osi po segmentima')
plt.xlabel('Segment Putanje')
plt.ylabel('Apsolutna pogreška u Y osi (cm)')
plt.grid(True)
plt.show()
```
